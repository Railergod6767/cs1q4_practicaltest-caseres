```python
def process_quiz_scores():
    # 1.
    student_name = input("Enter student's name: ")

    while True:
        try:
            # 2.
            num_scores = int(input("Enter how many scores you will input: "))
            if num_scores <= 0:
                print("Please enter a positive number of scores.")
            else:
                break
        except ValueError:
            print("Invalid input. Please enter an integer.")

    scores = []
    for i in range(num_scores):
        while True:
            try:
    # 3.
                score = float(input(f"Enter quiz score {i + 1}: "))
                if 0 <= score <= 100:
                    scores.append(score)
                    break
                else:
                    print("Score must be between 0 and 100.")
            except ValueError:
                print("Invalid input. Please enter a number.")

    # 4. functions
    total_score = sum(scores)
    average_score = total_score / num_scores if num_scores > 0 else 0

    # 5.
    passing_score = 75
    if average_score >= passing_score:
        status = "Passed"
    else:
        status = "Failed"

    print(f"\n--- Quiz Report for {student_name} ")
    print(f"Total scores inputted: {num_scores}")
    print(f"Individual scores: {scores}")
    print(f"Total Score: {total_score:.2f}")
    print(f"Average Score: {average_score:.2f}")
    print(f"Status: {status}")

process_quiz_scores()
```
